```python
import csv

# 打开输入文件和输出文件
with open('CO2_heat_pump_V12_7500__251103_batch_Midea4_1_cabin_1.csv', 'r') as input_file, open('output_1.csv', 'w', newline='') as output_file:
    # 创建CSV读取器和写入器
    reader = csv.reader(input_file, delimiter=';')
    writer = csv.writer(output_file, delimiter=',')

    # 对于每一行数据
    for i, row in enumerate(reader):
        # 第1到4行和第6行的数据都会被删除
        if i not in [0, 1, 2, 3, 5]:
            # 去除行首的#号，并将每个数据分割成多个部分
            parts = []
            for item in row:
                parts.extend([x.lstrip('#').strip() for x in item.split(';')])
            # 将拆分后的数据写入输出文件
            writer.writerow(parts)
```
